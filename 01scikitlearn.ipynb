### 손글씨를 이미지화해서 화면에 출력

#### 이미지 데이터와 레이블을 하나로 묶어서 리스트로 생성

```python
imglbl = list(zip(digits.images, digits.target))
```

#### 2x5 형태로 이미지와 레이블을 그림형태로 출력

```python
for ix, (img, lbl) in enumerate(imglbl[:10]):
    plt.subplot(2, 5, ix+1)
    plt.axis(False)
    plt.imshow(img, cmap=plt.cm.gray_r, interpolation='nearest')
    plt.title(f'trainig : {lbl}')
```

### 가상데이터 생성하기

#### 분류 테스트를 위해 make_blob, make_classification, make_moon, make_circle등을 이용해서 가상의 데이터를 생성할 수 있음

#### make_classification

```python
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(# n_samples : 표본수, 기본 : 100
                            n_features=2, # 독립변수 개수
                           n_informative=1, # x,y의 상관관계 여부
                           n_redundant=0, # x,y의 선형관계 여부
                           n_clusters_per_class=1, # 군집 개수
                           random_state=2007281430)
```

```python
plt.scatter(X[:,0], X[:, 1], marker='o', c=y, s=100, edgecolor='k')
```

### 정규분포를 통해 군집용 가상데이터 생성

```python
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_features=2, 
                  centers = 3, # 군집 중심점 개수
                  random_state=2111031705)
```

```python
plt.scatter(X[:, 0], X[:,1], marker='o', c=y, s=100, edgecolors='k')
```
