#### 간단한 sin 그래프

```python
x = np.linspace(0, 9.5, 1000)
```

```python
fig = plt.figure()
ax = plt.axes()  # 축 영역
ax.plot(x, np.sin(x), 'r')
ax.grid()
```

#### 간단한 cosine, tangent 그래프

```python
fig = plt.figure()
ax = plt.axes()  # 축 영역
ax.plot(x, np.cos(x), 'b')
ax.grid()
```

```python
fig = plt.figure()
ax = plt.axes()  # 축 영역
ax.plot(x, np.sin(x), 'r')
ax.plot(x, np.cos(x), 'b')
ax.plot(x, np.tan(x), 'g')
ax.set_xlim(0, 9.5)
ax.set_ylim(-5, 5)
ax.grid()
```

```python
fig = plt.figure()
ax = plt.axes()  # 축 영역
ax.plot(x, np.sin(x), 'r', label='사인')
ax.plot(x, np.cos(x), 'b', label='코사인')
ax.plot(x, np.tan(x), 'g', label='탄젠트')
ax.set(xlim=(0, 9.5), ylim=(-1.1, 1.1), title='sin/cos/tan 그래프', xlabel='x값', ylabel='y값')
ax.legend()
ax.grid()
```

#### 그래프로 시각화2


```python
fig = plt.figure()
ax = plt.axes()
plt.plot(x, y1, 'r', label = 'busan-seoul')
plt.plot(x, y2, 'b', label = 'busan-kyonggi')
plt.plot(x, y3, 'm', label = 'busan-incheon')
ax.legend()
# x/y축 눈금을 리스트로 변환후 포맷문자열로 변환
ax.set_xticklabels(['{:}'.format(int(x)) for x in ax.get_xticks().tolist()])
ax.set_yticklabels(['{:,}'.format(int(y)) for x in ax.get_yticks().tolist()])
plt.show()
```

#### 그래프로 시각화3

```python
plt.subplot(2,1,1)
plt.plot(x, y1, 'r')
plt.title('pusan -> seoul')
plt.xlim([2000, 2017])
plt.show()
```

```python
ax2 = plt.subplot(2,2,3)

plt.plot(x, y2, 'b')

plt.title('pusan->kyonggi')

plt.xlim(2000,2017)
```

```python
ax3 = plt.subplot(2, 2, 4, sharey=ax2)
# 이전 그래프의 y축을 공유
plt.plot(x, y2, 'm')
plt.title('pusan -> incheon')
plt.xlim(2000, 2017)
plt.yticks(visible=False) # 눈금표시안함
```
