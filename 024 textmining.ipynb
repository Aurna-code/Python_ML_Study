### 토큰화를 위한 패키지 설치

```python
from nltk import sent_tokenize
from nltk import word_tokenize
```

### 토큰화에 필요한 데이터 다운로드

```python
nltk.download('punkt')
```

```python
text_sample = 'The Matrix is everywhere its all around us, here even in this room. \
               You can see it out your window or on your television. \
               You feel it when you go to work, or go to church or pay your taxes.'
```

### 문장 토큰화

```python
sents = sent_tokenize(text_sample)
sents  # 결과는 리스트로 저장
```

### 단어 토큰화 : word_tokenize(문장)

```python
words = word_tokenize(sents[0])
words  # 결과는 리스트로 저장
```

### 문장 토큰화한 후 단어 토큰화 실시

```python
words = []
sents = sent_tokenize(text_sample)
for sent in sents: 
    words.append(word_tokenize(sent))
```

```python
words
```

```python
len(words)
```

### 불용어 stopword 제거
+ 분석에 큰 의미가 없는 단어를 지칭
+ NLTK상에서는 stopwords 형태로 제공됨
+ 하지만, 제공되는 단어수는 적음

```python
nltk.download('stopwords')
```

```python
nltk.corpus.stopwords.words('english')[:10] # 불용어들은 모두 소문자로 출력
```

### 텍스트 전처리후 단어 토큰화 실시

```python
stopwords = nltk.corpus.stopwords.words('english')
tokens = []
for wds in words:
    filter_words = []
    for w in wds:
        w = w.lower()  # 단어를 소문자로 바꾸기
        if w not in stopwords: 
            filter_words.append(w)
    tokens.append(filter_words)
```

```python
tokens
```

### 트럼프 연설문을 이용한 단어 토큰화

```python
with open('data/trump_en.txt') as f:
    docs = f.read()  # 텍스트파일을 통채로 읽어옴
```

```python
sents1 = sent_tokenize(docs)
```

```python
import re
```

```python
words1 = []
for s in sents1:
    # 알파벳 대소문자를 제외하고 모두 제거
    s = re.sub('[^a-zA-Z]',' ',s).strip()
    words1.append(word_tokenize(s))
words1[:1]
```

```python
tokens1 = []
for wds in words1:
    fwords =[]
    for w in wds:
        w = w.lower()
        if w not in stopwords:
            fwords.append(w)
    tokens1.append(fwords)
tokens1[:1]
```

```python
tokens2 = []
for wds in words2:
    fwords2 =[]
    for w in wds:
        w = w.lower()
        if w not in stopwords:
            fwords2.append(w)
    tokens2.append(fwords2)
tokens2[:1]
```

### 형태소 분석
+ 의미가 있는 최소의 단위로서 더 이상 분리가 불가능한 가장 작은 의미 요소
+ 주어진 단어 또는 어절을 구성하는 각 형태소를 분리한 후 분리된 형태소의 기본형 및 품사 정보를 추출하는 것
+ 문장 > 어절 > 단어 > 어근/접두사/접미사 > 품사
    - work : 동사/명사
    - working : working의 원형인 work를 추출 후 동사/명사
    - 영어의 경우 과거/현재/미래, 단수, 진행형 등에 따라 단어가 변화함
    - 보조적으로 어근추출 stemming이나 원형복원 lemmatizing이 필요할 수 있음

### 품사태깅 POS tagging
+ 단어를 문법적 기능이나 형태,의미에 따라 구분한 것
+ NLTK에서는 Penn Treebank Tagset을 이용해서 품사태깅을 함

```python
nltk.download('averaged_perception_tagger')
```

```python
from nltk.tag import pos_tag
```

### 기존 문장에 대해 품사태깅 실시

```python
tags = pos_tag(tokens[0])
```

```python
tags # (RB 부사, IN 종속 전치사, PRP대멍사)
```

### 분석에 필요한 토큰 추출 - 품사가 명사인 토큰을 대상

```python
nouns = []
```

```python
for t in tokens:
    tags = pos_tag(t)
    NN = [l[0] for l in tags if l[1] == 'NN']
    nouns.append(NN)
```

```python
nouns
```

```python
nouns1 = []
for t in tokens1:
    tags = pos_tag(t)
    NN = [ w[0] for w in tags if w[1] == 'NN']
    nouns1.append(NN)
nouns1[:5]
```

### 스티브 잡스 스탠포드대 졸업축사 품사 태깅

```python
nouns2 = []
for t in tokens2:
    tags = pos_tag(t)
    NN = [ w[0] for w in tags if w[1] == 'NN']
    nouns2.append(NN)
nouns2[:5]
```

### 단어빈도
+ 텍스트 데이터의 대표단어를 알아볼 수 있는 방법
+ 단어별 빈도를 조사해서 알아낸 상위 n개의 단어들을 대표단어로 지정

```python
from nltk import FreqDist
```

#### 먼저, 기존 단어들이 저장된 2차원 리스트를 1차원으로 바꿈

```python
NNO = []
for ns in nouns:
    for n in ns:
        NNO.append(n)
```

#### 단어별 빈도 파악

```python
frqwords = FreqDist(NNO)
frqwords
```

#### FreqDist 좀 더 살펴보기

```python
frqwords.N() # 7 : 전체 단어수
frqwords['room'] # 특정단어의 빈도수
frqwords.most_common() # 빈도수가 많은 순서대로 단어 출력
frqwords.most_common(3)
```

#### 트럼프 연설문에서 가장 많이 언급된 10개 단어

```python
NNO1 = []
for ns in nouns:
    for n in ns:
        NNO1.append(n)
```

```python
frqwords1 = FreqDist(NNO1)
frqwords1
```

```python
frqwords1.N() # 7 : 전체 단어수
frqwords1['room'] # 특정단어의 빈도수
frqwords1.most_common() # 빈도수가 많은 순서대로 단어 출력
frqwords1.most_common(10)
```

```python
NNO2 = []
for ns in nouns2:
    for n in ns:
        NNO2.append(n)
```

```python
frqwords2 = FreqDist(NNO2)
frqwords2
```

```python
frqwords2.N() # 7 : 전체 단어수
frqwords2['room'] # 특정단어의 빈도수
frqwords2.most_common() # 빈도수가 많은 순서대로 단어 출력
frqwords2.most_common(10)
```
